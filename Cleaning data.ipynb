```python
# Enlèvons 'Glassdor est.' de la colonne 'Salary Estimate'
df['Salary Estimate'] = df['Salary Estimate'].str.replace('Glassdoor est.', '')
df.head()
```

#### Séparons la colonne location en deux colonne 'City location' 'State location' 

```python
# En utilisant str.extract avec une expression régulière, nous pouvons avoir plus de flexibilité 
# pour capturer les motifs souhaités dans la colonne 'Location'. 
# L'expression régulière r'^(.*?),\s*(.*?)$' a été conçue pour capturer le contenu avant
# et après la virgule, tout en gérant les espaces potentiels.
df[['City location', 'State location']] = df['Location'].str.extract(r'^(.*?),\s*(.*?)$')
df.head(3)
```

```python
df = df.drop('Location', axis=1)
df.head(3)
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace(' ( )', '')
df.head(3)
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace('(', '')
df.head(3)
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace(')', '')
df.head(3)
```

```python
df[['Min Salary Estimate', 'Max Salary Estimate']] = df['Salary Estimate'].str.split('-', expand=True)
df.head(3)
```

```python
df['Min Salary Estimate'] = df['Min Salary Estimate'].str.replace('$', '')
df['Max Salary Estimate'] = df['Max Salary Estimate'].str.replace('$', '')
df['Min Salary Estimate'] = df['Min Salary Estimate'].str.replace('K', '')
df['Max Salary Estimate'] = df['Max Salary Estimate'].str.replace('K', '')
#df['Max Salary Estimate'] = df['Min Salary Estimate'].str.replace('$', '')
```

```python
df.head(3)
```

```python
df['Max Salary Estimate'] = df['Max Salary Estimate'].str.replace('Employer est.', '')
df.head(3)
```

```python
#df['Max Salary Estimate'] = df['Max Salary Estimate'].str.replace('(Employer est.)', '') 
#df.to_csv('job_v2.csv', sep=';',index=False)
```

```python
df['Min Salary Estimate'] = df['Min Salary Estimate'].fillna(0).astype(int) * 1000
df['Max Salary Estimate'] = df['Max Salary Estimate'].fillna(0).astype(int) * 1000
df[['Min Salary Estimate', 'Max Salary Estimate']].head()
```

```python
# Ici on utilise une expression régulière pour enlèver les chiffres de notre colonne 
df['Company Name'] = df['Company Name'].replace(to_replace='\d', value='', regex=True)
df.head()
```

```python
# Enlèvons le \n sur notre colonne
df['Company Name'] = df['Company Name'].str.replace('\n', '')
df['Company Name']
```

```python
# Enlèvons le \n sur notre colonne
df['Job Description'] = df['Job Description'].str.replace('\n', '')
df['Job Description']
```
